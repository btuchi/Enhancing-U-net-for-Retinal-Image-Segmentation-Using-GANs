```python
import torch
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
# 生成一些数据
x = np.linspace(0, 10, 10)
y = np.random.randint(1, 10, size=10)

# 使用plt.plot绘制线图
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(x, y, marker='o')
plt.title('Line Plot')

# 使用plt.step绘制阶梯图
plt.subplot(1, 2, 2)
plt.step(x, y, linestyle='--', marker='o')
plt.title('Step Plot')

plt.tight_layout()
plt.show()
```
